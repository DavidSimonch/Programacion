```python
data['Periodo'] = pd.to_datetime(data['Periodo'])
data.set_index('Periodo', inplace = True)
series = data['Alimentos y bebidas no alcohólicas']
```

```python
#Grafico de la serie temporal para Alimentos y bebidas.

plt.figure(figsize=(12, 6))
series.plot(title='Evolución de Alimentos y bebidas no alcohólicas (IPC)', xlabel='Periodo', ylabel='Índice')
plt.grid()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie temporal
decomposition = seasonal_decompose(series, model='additive', period=12)

# Graficar los componentes de la descomposición
plt.figure(figsize=(12, 10))
decomposition.plot()
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Prueba de Dickey-Fuller aumentada (ADF) para estacionariedad
adf_test = adfuller(series, autolag='AIC')

# Mostrar los resultados
adf_result = {
    'Estadístico ADF': adf_test[0],
    'p-valor': adf_test[1],
    'Lags usados': adf_test[2],
    'Número de observaciones': adf_test[3],
    'Valores críticos': adf_test[4]
}

adf_result
```

Los Valores no son estacionarios

```python
# Diferenciación de la serie
diff_series = series.diff().dropna()

# Prueba de estacionariedad nuevamente en la serie diferenciada
adf_test_diff = adfuller(diff_series, autolag='AIC')

# Resultados después de la diferenciación
adf_result_diff = {
    'Estadístico ADF': adf_test_diff[0],
    'p-valor': adf_test_diff[1],
    'Lags usados': adf_test_diff[2],
    'Número de observaciones': adf_test_diff[3],
    'Valores críticos': adf_test_diff[4]
}

# Graficar la serie diferenciada
plt.figure(figsize=(12, 6))
diff_series.plot(title='Serie Diferenciada: Alimentos y bebidas no alcohólicas', xlabel='Periodo', ylabel='Índice Diferenciado')
plt.grid()
plt.show()

adf_result_diff
```

Dado que el estadístico ADF está muy cerca del umbral crítico y el p-valor es razonablemente bajo, considero que la serie diferenciada está lo suficientemente cerca de la estacionariedad para proceder con el modelado SARIMAX.

Division de los conjuntos de entrenamiento y prueba

```python
train_size = int(len(diff_series) * 0.8)
train, test = diff_series[:train_size], diff_series[train_size:]

# Mostrar los tamaños de los conjuntos
len(train), len(test)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Graficar la ACF y PACF para la serie diferenciada
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_acf(diff_series, lags=20, ax=plt.gca(), title='ACF: Serie Diferenciada')

plt.subplot(1, 2, 2)
plot_pacf(diff_series, lags=20, ax=plt.gca(), title='PACF: Serie Diferenciada')
plt.tight_layout()
plt.show()
```

Esto sugiere que podríamos probar con:
	•	p = 1 o p = 2
	•	q = 1 o q = 2

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Entrenar el modelo SARIMAX con parámetros iniciales (p=1, d=1, q=1)
model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(0, 0, 0, 0), enforce_stationarity=False, enforce_invertibility=False)
results = model.fit(disp=False)

# Resumen del modelo
results.summary()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Predicciones para el conjunto de prueba
forecast_values = results.forecast(steps=len(test))

# Convertir las predicciones a una serie con los mismos índices que el conjunto de prueba
forecast_series = pd.Series(forecast_values, index=test.index)

# Métricas de error
mse = mean_squared_error(test, forecast_series)
mae = mean_absolute_error(test, forecast_series)

# Mostrar las métricas
print("Métricas de error:")
print(f"MSE (Error Cuadrático Medio): {mse}")
print(f"MAE (Error Absoluto Medio): {mae}")
```

```python
# Graficar las predicciones y los valores reales
plt.figure(figsize=(12, 6))
plt.plot(test.index, test, label='Valores reales', marker='o', linestyle='-', color='blue')
plt.plot(test.index, forecast_series, label='Predicciones', linestyle='--', color='orange')
plt.legend()
plt.title('Predicciones SARIMAX vs Valores Reales')
plt.xlabel('Periodo')
plt.ylabel('Índice Diferenciado')
plt.grid()
plt.show()
```

```python
# Reconstruir las predicciones en la escala original
last_train_value = series[train.index[-1]]  # Último valor del conjunto de entrenamiento
forecast_original_scale = forecast_series.cumsum() + last_train_value

# Graficar los valores originales y las predicciones reconstruidas
plt.figure(figsize=(12, 6))
plt.plot(series.index, series, label='Valores originales', color='blue', linestyle='-', marker='.')
plt.plot(forecast_original_scale.index, forecast_original_scale, label='Predicciones (Escala Original)', linestyle='--', color='orange')
plt.legend()
plt.title('Predicciones SARIMAX Reconstruidas vs Valores Originales')
plt.xlabel('Periodo')
plt.ylabel('Índice')
plt.grid()
plt.show()
```
